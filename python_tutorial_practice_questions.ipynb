f(100) > f(79) > f(45) > f(28) > f(27) > f(13) > f(10) > f(6) > f(5) > f(0)

```python
from random import uniform
from functools import reduce
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples


# you can free to change all these codes/structure
def pick_a_number_from_list(A):
    S=0.0
    for el in A:
        S += el
    B = [x/S for x in A]
    b=reduce(lambda c, x: c + [c[-1] + x], B, [0])[1:]
    r = uniform(0,1)
    for i in b:
        if r <= i:
            s = i
            n = b.index(s)
            return A[n]
    

def sampling_based_on_magnitued(A):
    for _ in range(0,100):
        number = pick_a_number_from_list(A)
        print(number)
A=[]

while True:
    a=input("Enter the elements of the matrix A : (To stop entering, enter N or n)")
    if a =='n' or a =='N':
        break
    A.append(int(a))
#A = [0, 5, 27, 6, 13, 28, 100, 45, 10, 79]
sampling_based_on_magnitued(A)
```

<h3> Q5: Find the closest points</h3>

Consider you are given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P

<br>Cosine distance between two points (x,y) and (p,q) is defined as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
from scipy import spatial

def closest_points_to_p(S, P):
    D = []
    for i in S:
        D.append(spatial.distance.cosine(i,P))
    m = sorted(range(len(D)), key = lambda k: D[k])
    for i in range(5):
        print(S[m[i]])

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
closest_points_to_p(S, P)
```

<pre>
Red =[(R11,R12),(R21,R22),(R31,R32),(R41,R42),(R51,R52),..,(Rn1,Rn2)]

<pre>

Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]

Output:
YES

NO

```python
def i_am_the_one(red,blue,line):
    a = line.split('x',1)
    b = a[1].split('y',1)
    x = float(a[0])
    y = float(b[0])
    c = abs(float(b[1]))
    re = []
    bl = []
    if (float(red[0][0])*x + float(red[0][1])*y)<c:
        r=0
    else:
        r=1
    for i in range(len(red)):
        if r==0:
            if (float(red[i][0])*x + float(red[i][1])*y)<c:
                re.append((float(red[i][0])*x + float(red[i][1])*y)<c)
        else:
            if (float(red[i][0])*x + float(red[i][1])*y)>c:
                re.append((float(red[i][0])*x + float(red[i][1])*y)>c)
    for i in range(len(blue)):
        if r==0:
            if (float(blue[i][0])*x + float(blue[i][1])*y)>c:
                bl.append((float(blue[i][0])*x + float(blue[i][1])*y)>c)
        else:
            if (float(blue[i][0])*x + float(blue[i][1])*y)<c:
                bl.append((float(blue[i][0])*x + float(blue[i][1])*y)<c)
    if len(re)==len(red) and len(bl)==len(blue):
        return "YES"
    else:
        return "NO"

Red= [(1,1),(2,1),(4,2),(2,4),(-1,4)]
Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
Lines=["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]

for i in Lines:
    yes_or_no = i_am_the_one(Red, Blue, i)
    print(yes_or_no) # the returned value
```

<h3> Q7: Filling the missing values in the specified format</h3>

<pre>

Ex 2: 40, _, _, _, 60 ==> (60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5 ==> 20, 20, 20, 20, 20 i.e. the sum of (60+40) is distributed qually to all 5 places<br>

    b. now distribute the sum (10+50) missing values in between (10, 10, 12, 12, 12, 12, 12, _, _) 
    c. now we will distribute 12 to right side missing values (10, 10, 12, 12, 12, 12, 4, 4, 4)
</pre>
for a given string with comma seprate values, which will have both missing values numbers like ex: "_, _, x, _, _, _"
you need fill the missing values

Q: your program reads a string like ex: "_, _, x, _, _, _" and returns the filled sequence

Ex:

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings


# you can free to change all these codes/structure
def first(a):
    di = int(a[0])
    uc=0
    for i in range(len(a)):
        if i == 0:
            continue
        if a[i] == '_':
            uc+=1
        else:
            if i == len(a)-1:
                di = di + int(a[i])
            else:
                b=first(a[:(i+1)])
                for j in range(i+1):
                    a[j]=b[j]
                c=first(a[i:])
                l=i
                for i in range(len(c)):
                    a[l]=c[i]
                    l+=1
                if a.count('_')==0:
                    return a
                    
                
    for i in range(len(a)):
        a[i] = di//len(a)
    return a

def last(a):
    uc=0
    for i in range(len(a)):
        if a[i] == '_':
            uc+=1
        else:
            di = int(a[i])
            for j in range(i+1):
                a[j] = di//len(a[:(i+1)])
            if i == len(a)-1:
                return a
            else:
                b=(first(a[i:]))
                l=i
                for i in range(len(b)):
                    a[l]=b[i]
                    l+=1
                return a

def curve_smoothing(S):
    a=S.split(',')
    if a[0] != '_':
        a=first(a)
    else:
        a=last(a)
    return a#list of values

S=  input("Enter the comma separated string : ")
print(S)
smoothed_values= curve_smoothing(S)
print(smoothed_values)
```

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings



# you can free to change all these codes/structure
from fractions import Fraction
def compute_conditional_probabilites(A):
    sd = dict()
    pd = dict()
    a = []
    e = ['S1', 'S2', 'S3']
    x = [('F1', 'S1'), ('F1', 'S2'), ('F1', 'S3'), ('F2', 'S1')\
         , ('F2', 'S2'), ('F2', 'S3'), ('F3', 'S1'), ('F3', 'S2')\
         , ('F3', 'S3'), ('F4', 'S1'), ('F4', 'S2'), ('F4', 'S3'),\
         ('F5', 'S1'), ('F5', 'S2'), ('F5', 'S3')]
    b = []
    for el in A:
        a.append(el[1])
        b.append(tuple(el))
    for el in e:
        sd[el] = a.count(el)
    for el in x:
        pd[el] = b.count(el)
    for el in x:
        print("P(F={}|S=={}) = ".format(el[0],el[1]),Fraction(pd[el],sd[el[1]]))
    
        
            
    

A = [['F1','S1'],['F2','S2'],['F3','S3'],['F1','S2']\
     ,['F2','S3'],['F3','S2'],['F2','S1'],['F4','S1'],['F4','S3'],['F5','S1']]
compute_conditional_probabilites(A)
```

<h3> Q10: Error Function</h3>

You will be given a list of lists, each sublist will be of length 2 i.e. [[x,y],[p,q],[l,m]..[r,s]]
consider its like a martrix of n rows and two columns

a. the first column Y will contain interger values <br>
b. the second column $Y_{score}$ will be having float values <br>
Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.44982
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings
import math

# you can free to change all these codes/structure
def compute_log_loss(A):
    y=0
    for i in range(len(A)):
        y+=(A[i][0]*math.log10(A[i][1]))+((1-A[i][0])*(math.log10(1-A[i][1])))
    loss=(-1)*(1/len(A))*y
    return loss

A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
loss = compute_log_loss(A)
print(loss)
```
